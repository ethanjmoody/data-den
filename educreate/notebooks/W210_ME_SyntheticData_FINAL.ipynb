##  Synthetic Data Generation

```python
# Function to create synthetic questions
def create_synthetic_question(context, sample_qa):

  system_content = """
  You are an expert and teacher in high school History covering a broad range of topics.
  You will be given a History topic and some example questions for students to explore and answer.
  You need to generate 8 different questions for the given topic.

  Your response with the 8 different questions should be in the following format:
  'generated question'
  'generated question'
  ...
  'generated question'

  You must generate no more and no less than 8 different questions.
  """

  response = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
          {"role": "system", "content": system_content},
          {"role": "user", "content": context},
          {"role": "assistant", "content": sample_qa[0]["question"]},
          {"role": "user", "content": context},
          {"role": "assistant", "content": sample_qa[1]["question"]},
          {"role": "user", "content": context}
          ])

  return response.choices[0].message.content

# Function to create synthetic answers
def create_synthetic_answer(context, sample_qa, question):

  system_content = """
  You are an expert and teacher in high school History covering a broad range of topics.
  You will be given a History topic and some example questions and answers.
  You need to generate an answer in less than 500 words for a given question in this context:
  """ + context

  response = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
          {"role": "system", "content": system_content},
          {"role": "user", "content": sample_qa[0]["question"]},
          {"role": "assistant", "content": sample_qa[0]["answer"]},
          {"role": "user", "content": sample_qa[1]["question"]},
          {"role": "assistant", "content": sample_qa[1]["answer"]},
          {"role": "user", "content": question}
          ])

  return response.choices[0].message.content
```

### Generate Questions

```python
# Generate synthetic questions
topics = [(context_cuban, sample_qs_cuban),
          (context_battle1066, sample_qs_battle1066),
          (context_coldwar, sample_qs_coldwar),
          (context_caesar, sample_qs_caesar),
          (context_americancw, sample_qs_americancw)]

synthetic_questions = {}

for topic in topics:
  syn_qs = create_synthetic_question(topic[0], topic[1])
  synthetic_questions[topic[0]] = syn_qs
```

```python
# Inspect generated questions
for key in synthetic_questions:
  gen_qs = synthetic_questions[key]
  gen_ind_qs = [qs.strip() for qs in gen_qs.split('\n') if qs.strip()]
  print(f"Questions for Context: {key}")
  print(" " * 75)
  for qs in gen_ind_qs:
    print(qs)
  print("-" * 75)
```

### Generate Answers

```python
# Generate synthetic answers for synthetic questions
synthetic_qa = {}
index = 0

for i, (key, value) in enumerate(synthetic_questions.items()):
  gen_ind_qs = [qs.strip() for qs in value.split('\n') if qs.strip()]
  for qs in gen_ind_qs:
    ans = create_synthetic_answer(key, topics[i][1], qs)
    synthetic_qa[index] = {"question": qs, "answer": ans}
    index += 1
```

### Add Sample Questions and Randomize to Create Training and Testing Datasets

```python
synthetic_qa.keys()
```

```python
synthetic_dataset = synthetic_qa.copy()

# Add original sample questions
index = list(synthetic_qa.keys())[-1] + 1

for topic in topics:
  for key, qa_pair in topic[1].items():
    question = qa_pair["question"]
    answer = qa_pair["answer"]
    synthetic_dataset[index] = {"question": question, "answer": answer}
    index += 1

print(synthetic_dataset)
```

```python
# Get all keys from the dataset
keys = list(synthetic_dataset.keys())

# Randomly select 10 keys for the test dataset
test_keys = random.sample(keys, 10)

# Create the test dataset
synthetic_test_dataset = {key: synthetic_dataset[key] for key in test_keys}

# Create the training dataset with the remaining keys
training_keys = set(keys) - set(test_keys)
synthetic_training_dataset = {key: synthetic_dataset[key] for key in training_keys}

# Output the test and training datasets
print("Synthetic Test Dataset Keys:", synthetic_test_dataset.keys())
print("Synthetic Training Dataset Keys:", synthetic_training_dataset.keys())
```
